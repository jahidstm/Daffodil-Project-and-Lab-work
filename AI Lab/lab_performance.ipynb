```python
modeTestScore1 = df['Age'].mode()[0]
modeTestScore2 = df['Gender'].mode()[0]
modeTestScore3 = df['Occupation'].mode()[0]
modeTestScore4 = df['Relationship Status'].mode()[0]
modeTestScore5 = df['Education'].mode()[0]
modeTestScore6 = df['Physical Health'].mode()[0]
modeTestScore7 = df['Fitness Satisfaction'].mode()[0]
modeTestScore8 = df['Closeness to Others'].mode()[0]
modeTestScore9 = df['Mental Health'].mode()[0]
modeTestScore10 = df['Illnesses'].mode()[0]
modeTestScore11 = df['Family Relationships'].mode()[0]
modeTestScore12 = df['Friend Relationships'].mode()[0]
modeTestScore13 = df['Joy Source'].mode()[0]
modeTestScore14 = df['Improvement Areas'].mode()[0]
modeTestScore15 = df['Support Network'].mode()[0]
modeTestScore16 = df['Weekly Happiness'].mode()[0]
modeTestScore17 = df['Job Satisfaction'].mode()[0]
modeTestScore18 = df['Goal Achievement'].mode()[0]
modeTestScore19 = df['Self Esteem'].mode()[0]
modeTestScore20 = df['Social Style'].mode()[0]
modeTestScore21 = df['Best Year'].mode()[0]
modeTestScore22 = df['Worst Year'].mode()[0]
modeTestScore23 = df['Happiness Score'].mode()[0]


df['Age'] = df['Age'].fillna(modeTestScore1)
df['Gender'] = df['Gender'].fillna(modeTestScore2)
df['Occupation'] = df['Occupation'].fillna(modeTestScore3)
df['Relationship Status'] = df['Relationship Status'].fillna(modeTestScore4)
df['Education'] = df['Education'].fillna(modeTestScore5)
df['Physical Health'] = df['Physical Health'].fillna(modeTestScore6)
df['Fitness Satisfaction'] = df['Fitness Satisfaction'].fillna(modeTestScore7)
df['Closeness to Others'] = df['Closeness to Others'].fillna(modeTestScore8)
df['Mental Health'] = df['Mental Health'].fillna(modeTestScore9)
df['Illnesses'] = df['Illnesses'].fillna(modeTestScore10)
df['Family Relationships'] = df['Family Relationships'].fillna(modeTestScore11)
df['Friend Relationships'] = df['Friend Relationships'].fillna(modeTestScore12)
df['Joy Source'] = df['Joy Source'].fillna(modeTestScore13)
df['Improvement Areas'] = df['Improvement Areas'].fillna(modeTestScore14)
df['Support Network'] = df['Support Network'].fillna(modeTestScore15)
df['Weekly Happiness'] = df['Weekly Happiness'].fillna(modeTestScore16)
df['Job Satisfaction'] = df['Job Satisfaction'].fillna(modeTestScore17)
df['Goal Achievement'] = df['Goal Achievement'].fillna(modeTestScore18)
df['Self Esteem'] = df['Self Esteem'].fillna(modeTestScore19)
df['Social Style'] = df['Social Style'].fillna(modeTestScore20)
df['Best Year'] = df['Best Year'].fillna(modeTestScore21)
df['Worst Year'] = df['Worst Year'].fillna(modeTestScore22)
df['Happiness Score'] = df['Happiness Score'].fillna(modeTestScore23)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

categorical_columns = [
    'Age',
    'Gender',
    'Occupation',
    'Relationship Status',
    'Education',
    'Physical Health',
    'Fitness Satisfaction',
    'Closeness to Others',
    'Mental Health',
    'Illnesses',
    'Family Relationships',
    'Friend Relationships',
    'Joy Source',
    'Improvement Areas',
    'Support Network',
    'Weekly Happiness',
    'Job Satisfaction',
    'Goal Achievement',
    'Self Esteem',
    'Social Style',
    'Best Year',
    'Worst Year',
    'Happiness Score'
]

for column in categorical_columns:
    if column in df.columns:
        df[column] = label_encoder.fit_transform(df[column].astype(str))
    else:
        print(f"Warning: '{column}' not found.")

print("Label Encoding Completed Successfully.")
df.head()
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder

categorical_columns = df.select_dtypes(include=['object']).columns
encoder = LabelEncoder()

for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col].astype(str))

target_col = 'Happiness Score'
X = df.drop(columns=[target_col])
y = df[target_col]

selector = SelectKBest(score_func=chi2, k=5)
X_new = selector.fit_transform(X, y)

selected_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_indices]

print("Selected Features for Happiness Prediction:")
print(selected_features.tolist())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

print("Training data shape (X_train, y_train):", X_train.shape, y_train.shape)
print("Testing data shape (X_test, y_test):", X_test.shape, y_test.shape)
```

NAIVE_BAYES MODEL

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = np.array(pd.crosstab(y_test, y_pred))
sns.heatmap(cm, annot=True, fmt=".2f")
plt.xlabel("True Label")
plt.ylabel("Predicted Label")
plt.title("Confusion Matrix for Naive Bayes Classifier")
plt.show()
```

DECESSION TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt=".2f")
plt.xlabel("True Label")
plt.ylabel("Predicted Label")
plt.title("Confusion Matrix for Decision Tree Classifier")
plt.show()
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
cm = np.array(pd.crosstab(y_test, y_pred))
sns.heatmap(cm, annot=True, fmt=".2f")
plt.xlabel("True Label")
plt.ylabel("Predicted Label")
plt.title("Confusion Matrix for KNN Classifier")
plt.show()
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = np.sum(y_pred == y_test) / len(y_test)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = np.array(pd.crosstab(y_test, y_pred))
sns.heatmap(cm, annot=True, fmt=".2f")

plt.xlabel("True Label")
plt.ylabel("Predicted Label")
plt.title("Confusion Matrix for Logistic Regression Classifier")
plt.show()
```

LINEAR REGRESSION

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("MAE:", mae)
print("MSE:", mse)
print("R^2:", r2)
```
